```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
from ipywidgets import interact

plt.style.use("seaborn-v0_8-whitegrid")
```

```python
months = [
    "Jan","Feb","Mar","Apr","May","Jun",
    "Jul","Aug","Sep","Oct","Nov","Dec"
]

np.random.seed(42)

data = {
    "Month": months,
    "Labor_Cost": np.random.randint(28000, 35000, 12),
    "Rent": [20000] * 12,
    "Utilities": np.random.randint(3000, 6000, 12),
    "Maintenance": np.random.randint(2000, 5000, 12),
    "Equipment_Lease": np.random.randint(4000, 7000, 12)
}

df = pd.DataFrame(data)
df["Total_Cost"] = df.iloc[:, 1:].sum(axis=1)
df
```

```python
def plot_monthly_trend(cost_type):
    plt.figure(figsize=(12,6))
    plt.plot(df["Month"], df[cost_type], marker="o", linewidth=2)
    plt.title(f"{cost_type.replace('_',' ')} Over Time")
    plt.xlabel("Month")
    plt.ylabel("Cost (€)")
    plt.grid(True)
    plt.show()

interact(
    plot_monthly_trend,
    cost_type=widgets.Dropdown(
        options=list(df.columns[1:-1]),
        description="Cost Type:"
    )
)
```

```python
def plot_breakdown(selected_month):
    row = df[df["Month"] == selected_month].iloc[0]
    categories = ["Labor_Cost","Rent","Utilities","Maintenance","Equipment_Lease"]
    values = row[categories].values

    plt.figure(figsize=(8,8))
    plt.pie(values, labels=categories, autopct="%1.1f%%", startangle=90)
    plt.title(f"Cost Breakdown — {selected_month}")
    plt.show()

interact(
    plot_breakdown,
    selected_month=months
)
```

```python
def simulate_labor_increase(percentage):
    df_sim = df.copy()
    df_sim["Labor_Adjusted"] = df_sim["Labor_Cost"] * (1 + percentage/100)
    df_sim["Total_Adjusted"] = (
        df_sim["Labor_Adjusted"] +
        df_sim["Rent"] +
        df_sim["Utilities"] +
        df_sim["Maintenance"] +
        df_sim["Equipment_Lease"]
    )

    plt.figure(figsize=(12,6))
    plt.plot(df_sim["Month"], df_sim["Total_Cost"], marker="o", label="Original")
    plt.plot(df_sim["Month"], df_sim["Total_Adjusted"], marker="o", label=f"Labor +{percentage}%")
    plt.title("Scenario Simulation — Labor Cost Increase")
    plt.xlabel("Month")
    plt.ylabel("Total Cost (€)")
    plt.legend()
    plt.grid(True)
    plt.show()

interact(
    simulate_labor_increase,
    percentage=widgets.IntSlider(min=0, max=30, step=1, value=10)
)
```

```python
df_plot = df.set_index("Month")[["Labor_Cost","Rent","Utilities","Maintenance","Equipment_Lease"]]

df_plot.plot(kind="bar", stacked=True, figsize=(14,7))
plt.title("Warehouse Monthly Cost Breakdown")
plt.ylabel("Cost (€)")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.set_index("Month").iloc[:, :-1], annot=True, cmap="Blues")
plt.title("Warehouse Cost Heatmap")
plt.show()
```

### 📊 Dashboard Insights

- Labor is the dominant cost driver across all months.
- Rent is stable and predictable, ideal for long-term planning.
- Utilities and maintenance show variability, suggesting efficiency opportunities.
- Scenario simulation highlights how sensitive total costs are to labor fluctuations.
- The dashboard format supports quick decision-making for warehouse managers and operations leaders.
